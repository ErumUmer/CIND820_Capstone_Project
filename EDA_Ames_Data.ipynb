```python
# Importing the libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Load the dataset
ames_data = pd.read_csv('AmesHousing.csv')
```

```python
# Display the first few rows of the dataset
print(ames_data.head())
```

```python
# Summary statistics of numerical features
print(ames_data.describe())
```

```python
# Data types and missing values
print(ames_data.info())
```

```python
#plot histograms for numerical features
ames_data.hist(bins=30, figsize=(15, 10))
plt.show()
```

```python
#Create a correlation heatmap to visualize feature relationships
corr_matrix = ames_data.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=True)
plt.show()
```

```python
#Explore Catagorical features using bar plots
categorical_features = ames_data.select_dtypes(include=['object'])
for col in categorical_features:
    sns.countplot(data=ames_data, x=col)
    plt.xticks(rotation=90)
    plt.show()
```

```python
# Visualize the distribution of the target variable
sns.histplot(data=ames_data, x='SalePrice', kde=True)
plt.show()
```

```python
# Univariate analysis for a single variable, e.g., 'SalePrice'

# Histogram and density plot
sns.histplot(data=ames_data, x='SalePrice', kde=True)
plt.title('Distribution of SalePrice')
plt.show()
```

```python
# Summary statistics
print(ames_data['SalePrice'].describe())
```

```python
# Correlation heat map between numrical columns and sales prices

ames_data_num = ames_data.select_dtypes(exclude='object').reset_index()

# Correlation Heatmap for Sale Price

ames_data_num.corr()[['SalePrice']].sort_values(by='SalePrice', ascending=False)
plt.figure(figsize=(16,14), dpi=500)
chart4 = sns.heatmap(ames_data_num.corr()[['SalePrice']].sort_values(by='SalePrice', ascending=False),
                     vmin=-1, vmax=1, cmap='BrBG',
                     )
chart4.set_title('Sale Price Correlation Heatmap', fontdict={'fontsize': 12},
                 pad=12)
plt.show()
```

```python
#Outlier detection in the top correlated features
for feature in ['SalePrice', 'Overall Qual', 'Gr Liv Area', 'Garage Cars', 'Garage Area', 'Total Bsmt SF']:
    Q1 = ames_data[feature].quantile(0.25)
    Q3 = ames_data[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = (ames_data[feature] < lower_bound) | (ames_data[feature] > upper_bound)
    outlier_data = ames_data[outliers]
    print(f'Potential outliers in {feature}: {len(outlier_data)}')
```

```python
#scatter plots of top correlate features
# Create a scatter plot for each of the top correlated features
top_correlated_features = ['Overall Qual', 'Gr Liv Area', 'Garage Cars', 'Garage Area', 'Total Bsmt SF']

for feature in top_correlated_features:
    sns.scatterplot(data=ames_data, x=feature, y='SalePrice')
    plt.title(f'Scatter Plot of {feature} vs. SalePrice')
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.show()
```
